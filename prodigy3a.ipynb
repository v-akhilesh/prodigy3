```python
import pandas as pd
import seaborn as sns
import plotly.express as plx
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
file1=pd.read_csv('/content/bank-full.csv',sep=';')
```

```python
file1.head(20)
```

```python
file2=pd.read_csv("bank-additional-full.csv",sep=";")
```

```python
file2.head(20)
```

```python
data = pd.concat([file1, file2], ignore_index=True)
print(data)
```

```python
count_values = {}
for col in data.columns:
    count_values[col] = len(data[col].unique())

for col, count in count_values.items():
    print(col, count)
```

```python
data.isnull().sum()
```

```python
data_1 = data.dropna(axis=1)
data_1
```

```python
data_1.isnull().sum()
```

```python
plx.pie(data_1,names="job")
```

```python
grouped = data_1.groupby('default').size().reset_index(name='num_employees')
plt.bar(grouped['default'], grouped['num_employees'], color=['red', 'blue'])
plt.xlabel('Default Status')
plt.ylabel('Number of Employees')
plt.title('Total Number of Employees by Default Status')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
